Реализуем autoencoder. Архитектуру (conv, fully-connected, ReLu, etc) можете выбирать сами. Экспериментируйте!

```python
import torch
import torch.nn as nn

class Autoencoder(nn.Module):
    def __init__(self):
        super().__init__()

        # ------- Encoder -------
        self.encoder = nn.Sequential(
            nn.Conv2d(3, 16, kernel_size=3, stride=2, padding=1),   # 45 → 23
            nn.ReLU(),
            nn.Conv2d(16, 8, kernel_size=3, stride=2, padding=1),   # 23 → 12
            nn.ReLU()
        )

        # ------- Decoder -------
        self.decoder = nn.Sequential(
            nn.ConvTranspose2d(8, 16, kernel_size=3, stride=2, padding=1, output_padding=1),  # 12 → 24
            nn.ReLU(),
            nn.ConvTranspose2d(16, 3, kernel_size=4, stride=2, padding=3, output_padding=1),  # 24 → 45
            nn.Sigmoid()
        )

    def forward(self, x):
        latent_code = self.encoder(x)
        reconstruction = self.decoder(latent_code)
        return reconstruction, latent_code
```

```python
criterion = nn.MSELoss()

autoencoder = Autoencoder()

optimizer = torch.optim.AdamW(autoencoder.parameters(), lr=0.001, weight_decay=0.01)

device = 'cuda:0' if torch.cuda.is_available() else 'cpu'
```

## 1.3 Обучение (2 балла)

Осталось написать код обучения автоэнкодера. При этом было бы неплохо в процессе иногда смотреть, как автоэнкодер реконструирует изображения на данном этапе обучения. Наример, после каждой эпохи (прогона train выборки через автоэнкодер) можно смотреть, какие реконструкции получились для каких-то изображений val выборки.

А, ну еще было бы неплохо выводить графики train и val лоссов в процессе тренировки =)

```python
n_epochs = 10
loss_history = []

for epoch in range(n_epochs):
    autoencoder.train()
    total_loss = 0

    for i, batch in enumerate(train_loader):
        # если batch — тензор, а не (data, label), то не распаковываем
        imgs = batch.to(device) if not isinstance(batch, (list, tuple)) else batch[0].to(device)

        # прямой проход
        outputs, _ = autoencoder(imgs)
        loss = criterion(outputs, imgs)

        # обратное распространение
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        total_loss += loss.item()

    avg_loss = total_loss / len(train_loader)
    loss_history.append(avg_loss)
    print(f"✅ Epoch [{epoch+1}/{n_epochs}], Avg Loss: {avg_loss:.4f}\n")

# -------- График --------
plt.figure(figsize=(8, 5))
plt.plot(range(1, n_epochs + 1), loss_history, marker='o')
plt.xlabel("Epoch")
plt.ylabel("Average Loss")
plt.title("Training Loss per Epoch")
plt.grid(True)
plt.show()
```

```python
def visualize_autoencoder(model, data_loader, device='cpu', num_images=6):
    model.eval()
    model.to(device)

    with torch.no_grad():
        for imgs in data_loader:
            imgs = imgs.to(device)
            reconstructions, latent_codes = model(imgs)
            break  # берем только первую партию

    imgs = imgs[:num_images].cpu().permute(0, 2, 3, 1).numpy()
    reconstructions = reconstructions[:num_images].cpu().permute(0, 2, 3, 1).numpy()
    latent_codes = latent_codes[:num_images].cpu()  # тензор с размером (num_images, channels, H, W)

    fig, axes = plt.subplots(3, num_images, figsize=(num_images*3, 9))

    for i in range(num_images):
        # Оригинал
        axes[0, i].imshow(imgs[i])
        axes[0, i].set_title("Original")
        axes[0, i].axis('off')

        # Реконструкция
        axes[1, i].imshow(reconstructions[i])
        axes[1, i].set_title("Reconstruction")
        axes[1, i].axis('off')

        # Латентный код: покажем по каналам первый из них (можно и больше, но так проще)
        latent_img = latent_codes[i, 0].numpy()
        axes[2, i].imshow(latent_img, cmap='viridis')
        axes[2, i].set_title("Latent Code (channel 0)")
        axes[2, i].axis('off')

    plt.tight_layout()
    plt.show()
```

__Подсказка:__ Если вместо лиц у вас выводится непонятно что, попробуйте посмотреть, как выглядят латентные векторы картинок из датасета. Так как в обучении нейронных сетей есть определенная доля рандома, векторы латентного слоя могут быть распределены НЕ как `np.random.randn(25, <latent_space_dim>)`. А чтобы у нас получались лица при запихивании вектора декодеру, вектор должен быть распределен так же, как латентные векторы реальных фоток.
```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
autoencoder = autoencoder.to(device)
autoencoder.eval()

# --- ШАГ 1: Получаем латенты из реальных картинок ---
latents_list = []

with torch.no_grad():
    for batch in train_loader:
        if isinstance(batch, (list, tuple)):
            imgs = batch[0].to(device)
        else:
            imgs = batch.to(device)
        _, latents = autoencoder(imgs)  # (B, 8, 12, 12)
        latents_list.append(latents.cpu())

latents_all = torch.cat(latents_list, dim=0)  # (N, 8, 12, 12)

# --- ШАГ 2: Вычисляем среднее и стандартное отклонение по латентам ---
mean = latents_all.mean(dim=0)  # (8, 12, 12)
std = latents_all.std(dim=0)    # (8, 12, 12)

# --- ШАГ 3: Генерируем новые латенты с таким же распределением ---
num_samples = 25
z = torch.normal(mean.expand(num_samples, -1, -1, -1),
                 std.expand(num_samples, -1, -1, -1)).to(device)

# --- ШАГ 4: Декодируем и визуализируем ---
with torch.no_grad():
    outputs = autoencoder.decoder(z)  # (25, 3, 45, 45)

# Приводим к [B, H, W, C] и визуализируем
outputs = outputs.cpu().permute(0, 2, 3, 1).numpy()

fig, axes = plt.subplots(5, 5, figsize=(10, 10))
for i, ax in enumerate(axes.flat):
    ax.imshow(outputs[i])
    ax.axis('off')
plt.suptitle("Generated Images from Latent Distribution")
plt.show()
```

## 2.1 Архитектура модели и обучение (2 балла)

Реализуем VAE. Архитектуру (conv, fully-connected, ReLu, etc) можете выбирать сами. Рекомендуем пользоваться более сложными моделями, чем та, что была на семинаре:) Экспериментируйте!

```python
features = 16
```

```python
class VAE(nn.Module):
    def __init__(self):
        super(VAE, self).__init__()


        self.features = features
        self.flatten = nn.Flatten()
        # encoder
        self.encoder = nn.Sequential(
            nn.Linear(in_features= 784, out_features= 400),
            nn.ReLU(),
            nn.Linear(in_features=400, out_features=2 * self.features) # удвоенный
        )

        # decoder
        self.decoder = nn.Sequential(
            nn.Linear(in_features=self.features, out_features=400),
            nn.ReLU(),
            nn.Linear(in_features=400, out_features=784)
        )

    def encode(self, x):
        x = self.flatten(x).float()
        x = self.encoder(x).view(-1, 2, self.features)
        x = x.view(-1, 2, self.features)

        mu = x[:, 0, :] # the first feature values as mean
        log_var = x[:, 1, :] # the other feature values as variance
        return mu, log_var

    def gaussian_sampler(self, mu, logsigma):
        if self.training:
            std = torch.exp(0.5 * logsigma)

            eps = torch.randn_like(std)

            z = mu + std * eps
            return z
        else:
            # на инференсе возвращаем не случайный вектор из нормального распределения, а центральный -- mu.
            # на инференсе выход автоэнкодера должен быть детерминирован.
            return mu

    def decode(self, z):
        x = self.decoder(z)
        reconstruction = torch.sigmoid(x)

        return reconstruction.view(-1, 1, 28, 28)

    def forward(self, x):
        mu, log_var = self.encode(x)
        z = self.gaussian_sampler(mu, log_var)
        reconstruction = self.decode(z)
        return reconstruction, mu, log_var
```

Определим лосс и его компоненты для VAE:

Надеюсь, вы уже прочитали материал в towardsdatascience (или еще где-то) про VAE и знаете, что лосс у VAE состоит из двух частей: KL и log-likelihood.

Общий лосс будет выглядеть так:

$$\mathcal{L} = -D_{KL}(q_{\phi}(z|x)||p(z)) + \log p_{\theta}(x|z)$$

Формула для KL-дивергенции:

$$D_{KL} = -\frac{1}{2}\sum_{i=1}^{dimZ}(1+log(\sigma_i^2)-\mu_i^2-\sigma_i^2)$$

В качестве log-likelihood возьмем привычную нам кросс-энтропию.

```python
import torch.nn.functional as F

def KL_divergence(mu, logsigma):
    return -0.5 * torch.sum(1 + logsigma - mu.pow(2) - logsigma.exp())

def log_likelihood(x, reconstruction):
    x = x.view(x.size(0), -1)
    reconstruction = reconstruction.view(reconstruction.size(0), -1)
    return F.binary_cross_entropy(reconstruction, x, reduction='sum')

def loss_vae(x, mu, logsigma, reconstruction):
    return KL_divergence(mu, logsigma) + log_likelihood(x, reconstruction)
```

И обучим модель:

```python
criterion = loss_vae

autoencoder = VAE()

optimizer = torch.optim.Adam(autoencoder.parameters(), lr=1e-3)

device = 'cuda:0' if torch.cuda.is_available() else 'cpu'
```

```python
n_epochs = 10
train_losses = []
val_losses = []

for epoch in tqdm(range(n_epochs)):
    autoencoder.train()
    train_loss_per = []

    for batch in train_loader:
        imgs = batch[0].to(device)
        optimizer.zero_grad()

        reconstruction, mu, logsigma = autoencoder(imgs)
        loss = loss_vae(imgs, mu, logsigma, reconstruction) / imgs.size(0)
        loss.backward()
        optimizer.step()
        train_loss_per.append(loss.item())

    train_losses.append(np.mean(train_loss_per))

    autoencoder.eval()
    val_loss_per = []
    with torch.no_grad():
        for batch in test_loader:
            imgs = batch[0].to(device)
            reconstruction, mu, logsigma = autoencoder(imgs)
            val_loss = loss_vae(imgs, mu, logsigma, reconstruction) / imgs.size(0)
            val_loss_per.append(val_loss.item())

    val_losses.append(np.mean(val_loss_per))

    print(f"Epoch {epoch+1}: Train Loss = {train_losses[-1]:.4f}, Val Loss = {val_losses[-1]:.4f}")

```

### Архитектура

На картинке ниже представлена архитектура простого Conditional VAE.

По сути, единственное отличие от обычного -- это то, что мы вместе с картинкой в первом слое энкодера и декодера передаем еще информацию о классе картинки.

То есть, в первый (входной) слой энкодера подается конкатенация картинки и информации о классе (например, вектора из девяти нулей и одной единицы). В первый слой декодера подается конкатенация латентного вектора и информации о классе.

На всякий случай: это VAE, то есть, latent у него все еще состоит из mu и sigma

Таким образом, при генерации новой рандомной картинки мы должны будем передать декодеру сконкатенированные латентный вектор и класс картинки.

P.S. Также можно передавать класс картинки не только в первый слой, но и в каждый слой сети. То есть на каждом слое конкатенировать выход из предыдущего слоя и информацию о классе.

```python
class CVAE(nn.Module):
    def __init__(self, num_classes = 10, features=16):
        super(CVAE, self).__init__()

        self.num_classes = num_classes
        self.flatten = nn.Flatten()
        self.features = features


        self.encoder = nn.Sequential(
            nn.Linear(28 * 28 + self.num_classes, 400),
            nn.ReLU(),
            nn.Linear(400, 2 * self.features)
        )

        self.decoder = nn.Sequential(
            nn.Linear(self.features + self.num_classes, 400),
            nn.ReLU(),
            nn.Linear(400, 28 * 28)
        )

    def encode(self, x, class_onehot):
        x = self.flatten(x)
        x = torch.cat([x, class_onehot], dim=1)
        x = self.encoder(x).view(-1, 2, self.features)

        mu = x[:, 0, :]
        logsigma =  x[:, 1, :]

        # <реализуйте forward проход энкодера
        # в качестве возвращаемых переменных -- mu, logsigma и класс картинки>

        return mu, logsigma

    def gaussian_sampler(self, mu, logsigma):
        if self.training:
            std = torch.exp(0.5 * logsigma)

            eps = torch.rand_like(std)

            return mu + eps * std
            # засемплируйте латентный вектор из нормального распределения с параметрами mu и sigma
        else:
            # на инференсе возвращаем не случайный вектор из нормального распределения, а центральный -- mu.
            # на инференсе выход автоэнкодера должен быть детерминирован.
            return mu

    def decode(self, z, class_onehot):
        z = torch.cat([z, class_onehot], dim = 1)
        reconstruction = self.decoder(z)
        reconstruction = torch.sigmoid(reconstruction).view(-1,1,28,28)

        # реализуйте forward проход декодера
        # в качестве возвращаемой переменной -- reconstruction

        return reconstruction

    def forward(self, x, class_onehot):
        mu, logsigma = self.encode(x, class_onehot)

        z = self.gaussian_sampler(mu, logsigma)
        reconstruction = self.decode(z, class_onehot)
        # используя encode и decode, реализуйте forward проход автоэнкодера
        # в качестве возвращаемых переменных -- mu, logsigma и reconstruction
        return mu, logsigma, reconstruction
```

```python
def cvae_loss(recon_x, x, mu, logvar):
    bce = F.binary_cross_entropy(recon_x, x, reduction='sum')
    kld = -0.5 * torch.sum(1 + logvar - mu.pow(2) - logvar.exp())
    return bce + kld
```

```python
cvae = CVAE(num_classes=10, features=16).to(device)
optimizer = torch.optim.Adam(cvae.parameters(), lr=1e-3)
epochs = 10

for epoch in range(epochs):
    cvae.train()
    total_loss = 0

    for batch_idx, (x, labels) in enumerate(train_loader):
        x = x.to(device)
        labels = labels.to(device)
        one_hot_labels = F.one_hot(labels, num_classes=10).float().to(device)

        optimizer.zero_grad()
        mu, logvar, recon = cvae(x, one_hot_labels)
        loss = cvae_loss(recon, x, mu, logvar)
        loss.backward()
        optimizer.step()

        total_loss += loss.item()

    print(f"Epoch {epoch+1}, Loss: {total_loss / len(train_loader.dataset):.4f}")
```
